```python
# Importing important libraries 
import urllib.request, sys, time
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
#news_count_threshold = 1000
```

```python
sub_cat_dict_story = {"business": ["industry", "economy", "markets"], 
                "sci-tech": ["health"], 
                "sports": ["football", "tennis", "motorsport"], 
                "entertainment": ["reviews"]}

sub_cat_dict_other = {"news": ["national", "international", "cities"],   #"states",
                "business": ["budget"],              #remember to skip quizzes
                "sci-tech": ["science", "technology", "internet"], 
                "sports": ["cricket"], 
                "entertainment": ["movies", "music", "art"]}
```

```python
#total_cat_count = len(sub_cat_dict)
#total_sub_cat_count = len([i for v in sub_cat_dict.values() for i in v])
#news_per_page_count = 15
#news_per_category = int(news_count_threshold/total_sub_cat_count)
```

```python
pages_per_sub_cat = 6
news_headlines = list()
news_links = list()
news_authors = list()
news_dates = list()
news_times = list()
news_entire = list()
news_cats = list()
news_subcats = list()
links = list()

for cat in sub_cat_dict_story:
    for sub_cat in sub_cat_dict_story[cat]:
        print(str(cat)+':'+str(sub_cat))
        for page_no in range(1, pages_per_sub_cat+1):
                # Specifying the URL 
                    url = "https://www.thehindu.com/{}/{}/?page={}".format(cat,sub_cat, page_no)
                    # Using BeautifulSoup for web crawling
                    page = requests.get(url)
                    soup = BeautifulSoup(page.content, "html.parser")
                    for i in soup.find_all('div',class_='story-card-news'):
                      head=i.h3.find('a',class_="").get_text()
                      if 'Quiz' in head:
                          continue
                      else:
                          news_links.append(i.h3.find('a',class_="").get('href'))
                          news_headlines.append(head)
                          links.append(i.h3.find('a',class_="").get('href'))
                          news_cats.append(cat)
                          news_subcats.append(sub_cat)
for cat in sub_cat_dict_other:
    for sub_cat in sub_cat_dict_other[cat]:
        print(str(cat)+':'+str(sub_cat))
        for page_no in range(1, pages_per_sub_cat+1):
                # Specifying the URL 
                    url = "https://www.thehindu.com/{}/{}/?page={}".format(cat,sub_cat, page_no)
                    # Using BeautifulSoup for web crawling
                    page = requests.get(url)
                    soup = BeautifulSoup(page.content, "html.parser")
                    try:
                      for i in soup.find('section', class_='otherArticles').find_all('div', class_="story-card-news"):
                        head=i.h3.find('a',class_="").get_text()
                        if 'Quiz' in head:
                          continue                          
                        else:
                          news_links.append(i.h3.find('a',class_="").get('href'))
                          news_headlines.append(head)
                          links.append(i.h3.find('a',class_="").get('href'))
                          news_cats.append(cat)
                          news_subcats.append(sub_cat)
                    except:
                      continue

                    for i in soup.find_all('div',class_="Other-StoryCard"):
                      head=i.h3.find('a',class_="").get_text()
                      if 'Quiz' in head:
                          continue
                      else:
                          news_links.append(i.h3.find('a',class_="").get('href'))
                          news_headlines.append(head)
                          links.append(i.h3.find('a',class_="").get('href'))
                          news_cats.append(cat)
                          news_subcats.append(sub_cat)
# iterating through links of page scraped from
for url1 in links:
                      page = requests.get(str(url1))
                      article = BeautifulSoup(page.content, "html.parser")
                      text =''
                      id1=article.find('article').get('data-id')
                      #print('content-body-'+id)
                      #print(article.find('div',id='content-body-'+id1))
                      for i in article.find('div',id='content-body-'+id1).find_all('p'):
                      #for i in article.find_all('p'):
                        text = text + ' ' + i.get_text()
                      try:
                        datetime = article.find('span', class_="blue-color ksl-time-stamp").get_text().split()
                        date = " ".join(datetime[0:3])
                        time = " ".join(datetime[3:5])
                      except:
                        date = 'date not available'
                        time = 'time not available'
                      #try:
                      #  time = article.time.attrs['datetime'].split('T')[1].split('+')[0]
                      #except:
                      #  time = 'time not available'
                      try:
                        author = article.find('span',class_='mobile-author').find('a', class_='mobile-auth-nm lnk').get_text()
                      except:
                        author = 'author not available'
                      
                      news_authors.append(author)
                      news_dates.append(date)
                      news_times.append(time)
                      news_entire.append(text)


                            
                            
                            


                
```

```python
News_Corpus = pd.DataFrame({"Date": news_dates, "Time": news_times, "Category": news_cats,
                            "Sub-Category": news_subcats, "Headline": news_headlines,
                           "Entire_News": news_entire, "Author": news_authors, "News_Link": news_links})
```

```python
News_Corpus['Headline'] = News_Corpus["Headline"].apply(lambda x:x.strip())
```
